```python
import os
import csv
import pandas as pd

with open('train.db', 'w', newline='') as f:
    data = pd.read_csv("base_classifier_train.csv") 
    #for i in range(0,len(data['Word'])):
    i=0
    while data['SentenceID'][i] <= 50:
        if len(data['Predicted'][i]) > 2 :
            string = "Etype("
            #if data['Gold Truth'][i] != "O" and data['Gold Truth'][i][2:] != "Other":
            string = string + str(data['SentenceID'][i]) + "," + str(data['tokenID'][i]) + "," + data['Predicted'][i][2:] + ")"
            print(string,file=f)
        else :
            string = "Etype("
            #if data['Gold Truth'][i] != "O" and data['Gold Truth'][i][2:] != "Other":
            string = string + str(data['SentenceID'][i]) + "," + str(data['tokenID'][i]) + ",Other" + ")"
            print(string,file=f)
        if len(data['Gold Truth'][i]) > 2 :
            string = "EFtype("
            #if data['Gold Truth'][i] != "O" and data['Gold Truth'][i][2:] != "Other":
            string = string + str(data['SentenceID'][i]) + "," + str(data['tokenID'][i]) + "," + data['Gold Truth'][i][2:] + ")"
            print(string,file=f)
        else:
            string = "EFtype("
            #if data['Gold Truth'][i] != "O" and data['Gold Truth'][i][2:] != "Other":
            string = string + str(data['SentenceID'][i]) + "," + str(data['tokenID'][i]) + ",Other" + ")"
            print(string,file=f)
        i = i+1
    data = pd.read_csv("relation_classifier_train.csv") 
    #for i in range(0,len(data['SentenceID'])):
    i=0
    while data['SentenceID'][i] <= 50:
        string = "Rtype("   
        #if data['Gold Truth'][i] != "None":
        string = string + str(data['SentenceID'][i]) + "," + str(data['token1'][i]) + "," +str(data['token2'][i]) + "," + data['Predicted'][i] + ")"
        print(string,file=f)
        string = "RFtype(" 
        #if data['Gold Truth'][i] != "None":
        string = string + str(data['SentenceID'][i]) + "," + str(data['token1'][i]) +"," +str(data['token2'][i])+ "," + data['Gold Truth'][i] + ")"
        print(string,file=f)
        i = i+1
```

```python
import os
import csv
import pandas as pd
with open('test.db', 'w', newline='') as f:
    data = pd.read_csv("base_classifier_test.csv") 
    for i in range(0,len(data['Word'])):
        string = "Etype("
        if data['Predicted'][i] != "O" and data['Predicted'][i][2:] != "Other":
            string = string + str(data['SentenceID'][i]) + "," + str(data['tokenID'][i]) + "," + data['Predicted'][i][2:] + ")"
            print(string,file=f)
    data = pd.read_csv("relation_classifier_test.csv") 
    for i in range(0,len(data['SentenceID'])):
        string = "Rtype("  
        if data['Predicted'][i] != "None":
            string = string + str(data['SentenceID'][i]) + "," + str(data['token1'][i]) + "," +str(data['token2'][i]) + "," + data['Predicted'][i] + ")"
            print(string,file=f)
```

```python
import os
import csv
import pandas as pd
data_entity = pd.read_csv("base_classifier_test.csv")
data_relation = pd.read_csv("relation_classifier_test1.csv") 
pointer_entity = 0
pointer_relation = 0
for j in range(1,289,1):
    name = "test" + str(j) + ".db"
    with open(name, 'w', newline='') as f:
        data = data_entity.loc[data_entity['SentenceID'] == j]
        print(data)
        print(len(data['Word']))
        for i in range(pointer_entity,pointer_entity+len(data['Word'])):
            string = "Etype("
            if data['Gold Truth'][i] != "O" and data['Gold Truth'][i][2:] != "Other":
                if len(data['Predicted'][i]) > 2:
                    string = string + str(data['SentenceID'][i]) + "," + str(data['tokenID'][i]) + "," + data['Predicted'][i][2:] + ")"
                    print(string,file=f)
                else:
                    string = string + str(data['SentenceID'][i]) + "," + str(data['tokenID'][i]) + ",Other" + ")"
                    print(string,file=f)
        pointer_entity = pointer_entity + len(data['Word'])
        data = data_relation.loc[data_relation['SentenceID'] == j]
        print(data)
        for i in range(pointer_relation,pointer_relation+len(data['SentenceID'])):
            #if we don't take data['Predicted'] = "None", we may miss on some Rtypes which are actually not none.
            if data['Gold Truth'][i] != "None":
                string = "Rtype(" 
                string = string + str(data['SentenceID'][i]) + "," + str(data['token1'][i]) + "," +str(data['token2'][i]) + "," + data['Predicted'][i] + ")"
                print(string,file=f)
        pointer_relation = pointer_relation + len(data['SentenceID'])
```
